```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error as MSE
from sklearn.model_selection import train_test_split
```

```python
cars_data = pd.read_csv("auto-mpg.csv")
```

```python
cars_data.head()
```

```python
cars_data= cars_data.drop(['car name'],axis=1)
```

```python
cars_data = cars_data.apply(lambda x : x.replace('?', np.nan))
cars_data = cars_data.dropna()
```

```python
X = cars_data.drop(['mpg'], axis=1).values
y= cars_data.mpg.values
```

```python
X_train,X_test, y_train,y_test = train_test_split(X,y, test_size=0.2, random_state=1)
```

# Decigion_Tree_Regressor_Model

```python
dt =  DecisionTreeRegressor(max_depth=8)
```

```python
dt.fit(X_train, y_train)
```

```python
y_pred = dt.predict(X_test)

y_pred[:5]
```

```python
mse = MSE(y_test, y_pred)

rmse = mse ** (1/2)
```

```python
print(mse)
print(rmse)
```
